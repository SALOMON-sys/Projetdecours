```python
# PARTIE 1 (INITIALISATION DU JEU)

#Importer les bibliothèques nécessaires
import tkinter as tk
from tkinter import messagebox

#Créer la fenêtre principale
root = tk.Tk()
root.title("TIC-TAC-TOE")

#Créer la liste des positions disponibles (1 à 9)
chiffres = [1, 2, 3, 4, 5, 6, 7, 8, 9]

#Initialiser la variable mark (X ou O)
mark = ''

#Variable pour compter le nombre de coups joués
count = 0

#Créer la liste des panneaux (10 cases, le premier est un espace vide)
panneaux = ["panneau"] + ["" for _ in range(9)]


#PARTIE 2 (VERIFICATION DU GAGNANT)
# Fonction pour vérifier la victoire
def win(player):
    return ((panneaux[1] == panneaux[2] == panneaux[3] == player) or
            (panneaux[4] == panneaux[5] == panneaux[6] == player) or
            (panneaux[7] == panneaux[8] == panneaux[9] == player) or
            (panneaux[1] == panneaux[4] == panneaux[7] == player) or
            (panneaux[2] == panneaux[5] == panneaux[8] == player) or
            (panneaux[3] == panneaux[6] == panneaux[9] == player) or
            (panneaux[1] == panneaux[5] == panneaux[9] == player) or
            (panneaux[3] == panneaux[5] == panneaux[7] == player))

# Fonction de vérification et mise à jour du jeu
def check(digit):
    global count, mark, chiffres

    if digit in chiffres:
        chiffres.remove(digit)
        count += 1

        # Déterminer le joueur (X pour joueur 1, O pour joueur 2)
        if count % 2 == 0:
            mark = 'O'
        else:
            mark = 'X'

        # Mettre à jour le panneau et le bouton
        panneaux[digit] = mark
        boutons[digit-1].config(text=mark, state="disabled")

        # Vérifier la victoire
        if win(mark):
            if mark == 'X':
                messagebox.showinfo("Résultat", "Le joueur 1 gagne !")
            else:
                messagebox.showinfo("Résultat", "Le joueur 2 gagne !")
            root.destroy()

        # Vérifier égalité
        elif count > 8:
            messagebox.showinfo("Résultat", "Match à égalité !")
            root.destroy()
            
# Création du plateau de jeu
boutons = []
for i in range(1, 10):
    b = tk.Button(root, text=" ", font=('Arial', 20), width=6, height=3,
                  command=lambda i=i: check(i))
    b.grid(row=(i-1)//3, column=(i-1)%3)
    boutons.append(b)

root.mainloop()
```
